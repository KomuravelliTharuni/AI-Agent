```python
!pip install transformers accelerate bitsandbytes fpdf

user_details = {
    "name": input("Enter your full name: "),
    "email": input("Enter your email: "),
    "phone": input("Enter your phone number: "),
    "skills": input("List your top skills (comma-separated): "),
    "tools": input("List the tools/technologies you know: "),
    "education": input("Enter your education details: "),
    "projects": input("Briefly describe one major project: "),
}


from transformers import pipeline

# Load a small open-source model for resume generation
resume_writer = pipeline("text-generation", model="tiiuae/falcon-rw-1b")

# You can switch model if falcon is slow
# resume_writer = pipeline("text-generation", model="google/flan-t5-small")

prompt = f"""
Create a professional resume for:
Name: {user_details['name']}
Phone: {user_details['phone']}
Email: {user_details['email']}
Skills: {user_details['skills']}
Tools: {user_details['tools']}
Education: {user_details['education']}
Projects: {user_details['projects']}

Focus on Data Analyst role with 3 years of experience.
"""
result = resume_writer(prompt, max_length=500, do_sample=True)
generated_resume = result[0]['generated_text']
print(generated_resume)


from fpdf import FPDF

pdf = FPDF()
pdf.add_page()
pdf.set_font("Arial", size=12)

# Split resume text into lines for PDF formatting
lines = generated_resume.split('\n')
for line in lines:
    pdf.multi_cell(0, 10, txt=line)

# Save file
pdf_filename = f"{user_details['name'].replace(' ', '_')}_Resume.pdf"
pdf.output(pdf_filename)

print(f"✅ PDF saved as: {pdf_filename}")


import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart
from email.mime.application import MIMEApplication

sender_email = "tharunikomuravelli@gmail.com"  # Use app password if using Gmail
receiver_email = user_details["email"]
password = "aqux eesd pqvn qdpg"  # NEVER use your actual Gmail password directly

# Email content
msg = MIMEMultipart()
msg["Subject"] = "Your Auto-Generated Resume"
msg["From"] = sender_email
msg["To"] = receiver_email

body = MIMEText("Please find attached your resume.")
msg.attach(body)

# Attach PDF
with open(pdf_filename, "rb") as f:
    part = MIMEApplication(f.read(), Name=pdf_filename)
    part['Content-Disposition'] = f'attachment; filename="{pdf_filename}"'
    msg.attach(part)

# Send email
with smtplib.SMTP_SSL("smtp.gmail.com", 465) as server:
    server.login(sender_email, password)
    server.send_message(msg)

print("📧 Email sent successfully!")

```
